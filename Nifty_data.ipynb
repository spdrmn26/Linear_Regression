```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
data = pd.read_csv('Nifty_corrected.csv')
data.head()
```

```python
data['Close'] = data['Close'].shift(-1)
data.tail()
```

```python
data.drop(246, inplace = True)
data.tail()
```

```python
y = data.Close
x = data.drop('Close', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
x_test.iloc[0]
```

```python
new_row = {'Date':246, 'Open':22248.85, 'High':22249.4, 'Low':21997.95, 'Shares Traded':364545679, 'Turnover (₹ Cr)':29741.39}
x_test = x_test.append(new_row, ignore_index=True)
```

```python
y_pred = model.predict(x_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
new_params = {
    'Date': 246,
    'Open': 22081.55,
    'High': 22252.50,
    'Low' : 21875.25,
    'Shares Traded': 343495136,
    'Turnover (₹ Cr)': 33985.94
}
```

```python
new_params_series = pd.Series(new_params)
```

```python
predicted_value = model.coef_.dot(new_params_series) + model.intercept_
print(predicted_value)
```
